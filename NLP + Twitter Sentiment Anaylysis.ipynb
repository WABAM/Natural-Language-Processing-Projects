```python
import pandas as pd
from textblob import TextBlob
import numpy as np
import re
```

```python
new_file = pd.read_csv('2015-09-02-senti.csv')
pos_sent = []
neu_sent = []
neg_sent = []
```

```python
new_file
```

```python
#https://docs.python.org/3/library/re.html
def clean_tweet(tweet):
    return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tweet).split())
```

```python
def analize_sentiment(tweet):
    analysis = TextBlob(clean_tweet(tweet))
    print(analysis)
    if analysis.sentiment.polarity > 0:
        pos_sent.append(analysis)
        return 1
    elif analysis.sentiment.polarity == 0:
        neu_sent.append(analysis)
        return 0
    else:
        neg_sent.append(analysis)
        return -1
```

```python
df = pd.DataFrame(np.array([analize_sentiment(tweet) for tweet in new_file['Text']]), columns=['SA'])
df.to_csv ('Sentiment.csv')
```

```python
print('total tweet : ', len(new_file))
print('Positive : ', len(pos_sent))
print('Neutral : ', len(neu_sent))
print('Negative : ', len(neg_sent))
```

```python
total_tweet = 'Total tweet : ' + str(len(new_file))
pos_twt = 'Positive : ' + str(len(pos_sent))
neu_twt = 'Neutral : ' + str(len(neu_sent))
neg_twt = 'Negative : ' + str(len(neg_sent))
pos_twt_per = 'Positive : ' + str(round(len(pos_sent) / len(new_file) * 100)) + '%'
neu_twt_per = 'Neutral : ' + str(round(len(neu_sent) / len(new_file) * 100)) + '%'
neg_twt_per = 'Negative : ' + str(round(len(neg_sent) / len(new_file) * 100)) + '%'
result = [total_tweet, pos_twt, neu_twt, neg_twt, pos_twt_per, neu_twt_per, neg_twt_per]
print(result)
```
